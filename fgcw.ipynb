```python
from collections import deque

# Define the initial state of the problem
# State is represented as a tuple (farmer, goat, cabbage, wolf)
# Where each element is either 'left' or 'right'
initial_state = ('left', 'left', 'left', 'left')
goal_state = ('right', 'right', 'right', 'right')

def is_valid_state(state):
    farmer, goat, cabbage, wolf = state
    # Check if the goat and cabbage are left alone on one side and wolf is on the same side as the goat
    if (goat == cabbage and wolf != goat) or (goat == wolf and cabbage != goat):
        return False
    return True

def get_possible_moves(state):
    farmer, goat, cabbage, wolf = state
    moves = []
    
    # Define the possible moves: 
    # 1. Farmer alone
    # 2. Farmer with goat
    # 3. Farmer with cabbage
    # 4. Farmer with wolf
    possible_moves = [
        ('farmer', 'none', 'none', 'none'),
        ('farmer', 'goat', 'none', 'none'),
        ('farmer', 'none', 'cabbage', 'none'),
        ('farmer', 'none', 'none', 'wolf')
    ]
    
    for move in possible_moves:
        new_farmer = 'right' if farmer == 'left' else 'left'
        new_goat = goat
        new_cabbage = cabbage
        new_wolf = wolf
        
        if move[0] == 'farmer':
            new_farmer = 'right' if farmer == 'left' else 'left'
        if move[1] == 'goat' and farmer == goat:
            new_goat = new_farmer
        if move[2] == 'cabbage' and farmer == cabbage:
            new_cabbage = new_farmer
        if move[3] == 'wolf' and farmer == wolf:
            new_wolf = new_farmer
            
        new_state = (new_farmer, new_goat, new_cabbage, new_wolf)
        if is_valid_state(new_state):
            moves.append(new_state)
    
    return moves

def solve_puzzle():
    queue = deque([(initial_state, [])])
    visited = set()
    
    while queue:
        (current_state, path) = queue.popleft()
        
        if current_state == goal_state:
            return path
        
        for next_state in get_possible_moves(current_state):
            if next_state not in visited:
                visited.add(next_state)
                queue.append((next_state, path + [next_state]))
    
    return None

def print_solution(solution):
    state_names = ['farmer', 'goat', 'cabbage', 'wolf']
    
    for step in solution:
        print('State:', ' '.join(f'{name} on the {side}' for name, side in zip(state_names, step)))

solution = solve_puzzle()
if solution:
    print_solution(solution)
else:
    print("No solution found")
```
